### Cleaning Data
Observations:
1. We strip the whitespaces as we observe more than expected unique values of nominal data. This fixes the observations.
2. We replace missing values with NaN
3. We convert the numeric columns's type from object to float64. 
4. We remove the null values using kNN imputation.

Strip the whitespaces

```python
for i in data_cols:
  data[i] = data[i].str.strip() 
```

Replace missing values with NaN

```python
data = data.replace('?', np.nan)
```

Converting string data into numbers 

```python
for i in data_numeric:
  data[i] = pd.to_numeric(data[i])
```

### Removing null values

**The process of removal of null values is as follows - <br>**

1. We convert the nominal data into numerical data next. The string nominal data is converted into numerical data. (Ternary - 0/1/missing)<br>
2. We convert all columns from type object to type float64.<br>
3. The null values of all numerical attributes are imputed using kNN imputation.<br>
* We are taking nominal attributes - specific gravity, albumin and sugar as numerical, i.e. the missing values are imputed using kNN imputation. <br>
* The null values of other nominal attributes - rbc, pc, wc, rc - are also 
imputed using kNN imputation. <br>
* rbc and pc are imputed by filling null values with median value to keep them nominal.



#### 1. Converting nominal data to numerical data


```python
def convert_to_numerical(attribute, positive, negative):
  global data
  data[attribute] = data[attribute].map({positive : 1, negative: 0, np.NaN : np.NaN})
```

```python
convert_to_numerical('rbc','normal','abnormal')
convert_to_numerical('pc','abnormal','normal')
convert_to_numerical('pcc','notpresent','present')
convert_to_numerical('ba','notpresent','present')
convert_to_numerical('htn','no','yes')
convert_to_numerical('dm','no','yes')
convert_to_numerical('cad','no','yes')
convert_to_numerical('appet','poor','good')
convert_to_numerical('pe','no','yes')
convert_to_numerical('ane','no','yes')
convert_to_numerical('class','notckd','ckd')

print(data.head(20))
```

```python
data.info()
```

#### 3. kNN Imputation.

```python
data['rbc'].fillna(data['rbc'].median(), inplace=True)
data['pc'].fillna(data['rbc'].median(), inplace=True)
```

```python
from sklearn.impute import KNNImputer
```

```python
imputer = KNNImputer(n_neighbors=2)
data = pd.DataFrame(imputer.fit_transform(data))
```

```python
column_names = ['age', 'bp', 'sg', 'al', 'su', 'rbc', 'pc', 'pcc', 'ba', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hemo', 'pcv', 'wc', 'rc', 'htn', 'dm', 'cad','appet', 'pe', 'ane', 'classification']
data.columns = column_names
```

```python
# Replace 0.5 with the median value
def replace_for_nominal(s):
  global data
  median = data['rbc'].median()
  data[s] = data[s].map({0.5 : median, 1 : 1, 0 : 0})
```

```python
replace_for_nominal('rbc')
replace_for_nominal('pc')
replace_for_nominal('pcc')
replace_for_nominal('ba')
replace_for_nominal('htn')
replace_for_nominal('dm')
replace_for_nominal('cad')
replace_for_nominal('appet')
replace_for_nominal('pe')
replace_for_nominal('ane')
```

## Exploring Skewness
15 attributes are highly skewed, 7 are moderately skewed and 3 are fairly symmetric.

```python
data.hist(figsize=(16, 20), bins=50, xlabelsize=8, ylabelsize=8)
```

**Skewness**

```python
skewness = data.skew(axis = 0)

highly_skewed = []
moderately_skewed = []
fairly_symmetrical = []

for item in skewness.iteritems():
  i = item[1]
  if abs(i) >= 1:
    highly_skewed.append(item)
  elif abs(i) < 1 and abs(i) >= 0.5:
    moderately_skewed.append(item)
  else:
    fairly_symmetrical.append(item)

def print_skewness(l, s):
  print(s, ' - ', len(l))
  for i in l:
    print(i[0], ' - ', round(i[1], 3))
  print("\n")

print_skewness(highly_skewed,'Highly Skewed')
print_skewness(moderately_skewed, 'Moderately Skewed')
print_skewness(fairly_symmetrical, 'Fairly Symmetrical')
```
