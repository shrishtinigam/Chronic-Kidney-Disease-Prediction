DATASET 1 LOGISTIC REGRESSION

```python
import pandas as pd           #chronic kidney disease
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df=pd.read_csv('/content/new_model.csv')
df
```

```python
X=df.drop(columns='Class',axis=1)
Y=df['Class']
# Normalizing
scaler=StandardScaler()
standardized_data=scaler.fit_transform(X)
print(standardized_data)
# Splitting the data into train and test
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2,random_state =42)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(X_train,Y_train)
print(model.score(X_test,Y_test))
predicts=model.predict(X_test)
```

```python
from sklearn.metrics import precision_score,recall_score,accuracy_score
print('Precision: %.3f' % precision_score(Y_test,predicts))
print('Recall: %.3f' % recall_score(Y_test, predicts))
print('Accuracy: %.3f' % accuracy_score(Y_test, predicts))
```
